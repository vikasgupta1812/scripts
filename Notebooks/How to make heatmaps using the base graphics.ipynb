```r
nba = nba[order(nba$PTS), ]
```

```r
row.names(nba) = nba$Name
```

```r
nba = nba[,2:20]
```

```r
head(nba)
```

```r
nba_matrix = data.matrix(nba)
```

```r
nba_heatmap = heatmap(nba_matrix, Rowv = NA, Colv = NA, 
                      col=cm.colors(256),
                     scale="column", margins=c(5,10))
```

```r
nba_heatmap = heatmap(nba_matrix, Rowv = NA, Colv = NA, 
                      col=heat.colors(256),
                     scale="column", margins=c(5,10))
```
