```r
# Create a fake data set
X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,1.98,1.98,2), 
                                         nrow=2, ncol=2))
plot(X)
```

```r
XX<-cbind(1,X)
b<-c(1,2,2)
y<-XX %*% b + rnorm(50, mean=0, sd=1)

model<-lm(y~X)
summary(model)
```

```r
# lower the level of multicollinearity

# Create a fake data set
X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,1.5,1.5,2), nrow=2, ncol=2))
plot(X)

XX<-cbind(1,X)
b<-c(1,2,2)
y<-XX%*%b+rnorm(50, mean=0, sd=1)

model<-lm(y~X)
summary(model)


# lower the level of multicollinearity again

# Create a fake data set
X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,1.0,1.0,2), nrow=2, ncol=2))
plot(X)

XX<-cbind(1,X)
b<-c(1,2,2)
y<-XX%*%b+rnorm(50, mean=0, sd=1)

model<-lm(y~X)
summary(model)


# set multicollinearity at zero

# Create a fake data set
X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,0,0,2), nrow=2, ncol=2))
plot(X)

XX<-cbind(1,X)
b<-c(1,2,2)
y<-XX%*%b+rnorm(50, mean=0, sd=1)

model<-lm(y~X)
summary(model)




# run a simulation to determine whether estimated standard errors are accurate

pb<-txtProgressBar(min=0, max=1000, style=3)
accuracy<-c()
falseneg<-c()
for(i in 1:1000){

  setTxtProgressBar(pb,i)
  # Create a fake data set
  X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,1.98,1.98,2), nrow=2, ncol=2))
  #plot(X)
  
  XX<-cbind(1,X)
  b<-c(1,2,2)
  y<-XX%*%b+rnorm(50, mean=0, sd=1)
  
  model<-lm(y~X)
  accuracy[i]<-ifelse((confint(model)[2,1]<2 & confint(model)[2,2]>2),1,0)
  falseneg[i]<-ifelse(summary(model)$coefficients[2,4]>0.05, 1, 0)
  
}

close(pb)
cat("% time that 95% CI covers true beta: ", 100*sum(accuracy)/1000, "\n")
cat("% false negatives: ", 100*sum(falseneg)/1000, "\n")


# run a simulation to determine whether estimated standard errors are accurate
# no multicollinearity

pb<-txtProgressBar(min=0, max=1000, style=3)
accuracy<-c()
falseneg<-c()
for(i in 1:1000){

  setTxtProgressBar(pb,i)
  # Create a fake data set
  X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,0,0,2), nrow=2, ncol=2))
  #plot(X)
  
  XX<-cbind(1,X)
  b<-c(1,2,2)
  y<-XX%*%b+rnorm(50, mean=0, sd=1)
  
  model<-lm(y~X)
  accuracy[i]<-ifelse((confint(model)[2,1]<2 & confint(model)[2,2]>2),1,0)
  falseneg[i]<-ifelse(summary(model)$coefficients[2,4]>0.05, 1, 0)
  
}

close(pb)
cat("% time that 95% CI covers true beta: ", 100*sum(accuracy)/1000, "\n")
cat("% false negatives: ", 100*sum(falseneg)/1000, "\n")















# Omitted Variable Bias
######################



# Create a fake data set
X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,1.5,1.5,2), nrow=2, ncol=2))
plot(X)
cor(X)

XX<-cbind(1,X)
b<-c(1,2,2)
y<-XX%*%b+rnorm(50, mean=0, sd=1)

model<-lm(y~X[,1])
summary(model)


# Do it again, but with only weakly correlated variables
# Create a fake data set

X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,0.25,0.25,2), nrow=2, ncol=2))
plot(X)
cor(X)

XX<-cbind(1,X)
b<-c(1,2,2)
y<-XX%*%b+rnorm(50, mean=0, sd=1)

model<-lm(y~X[,1])
summary(model)




# Do it again, but with totally uncorrelated
# Create a fake data set

X<-rmvnorm(50, mean=c(0,0), sigma=matrix(data=c(2,0,0,2), nrow=2, ncol=2))
plot(X)

XX<-cbind(1,X)
b<-c(1,2,2)
y<-XX%*%b+rnorm(50, mean=0, sd=1)

model<-lm(y~X[,1])
summary(model)




# Heteroskedasticity
######################

rm(list=ls())

# Create a fake data set
X<-runif(200, min=-10, max=10)
u<-rnorm(200, mean=0, sd=.2*(X+11))
y<-3+2*X+u


# run a standard model
model<-lm(y~X)
summary(model)
plot(model$residuals~X)


# show that 95% CIs are no longer accurate

pb<-txtProgressBar(min=0, max=1000, style=3)
accuracy<-c()
falseneg<-c()
for(i in 1:1000){

  setTxtProgressBar(pb,i)
  # Create a fake data set

  X<-runif(200, min=-10, max=10)
  u<-rnorm(200, mean=0, sd=.2*(X+11))
  y<-3+2*X+u

  model<-lm(y~X)
  accuracy[i]<-ifelse((confint(model)[2,1]<2 & confint(model)[2,2]>2),1,0)
  falseneg[i]<-ifelse(summary(model)$coefficients[2,4]>0.05, 1, 0)
  
}

close(pb)
cat("% time that 95% CI covers true beta: ", 100*sum(accuracy)/1000, "\n")
cat("% false negatives: ", 100*sum(falseneg)/1000, "\n")




# Create a fake data set
X<-runif(200, min=-10, max=10)
u<-rnorm(200, mean=0, sd=.2*(X+11))
y<-3+2*X+u


# run a standard model
model<-lm(y~X)
summary(model)
plot(model$residuals~X)


# now, use White's heteroskedasticity-robust standard errors
robust<-diag(sqrt(hccm(model, type="hc0")))
robust

# t-test
model$coefficients/robust

# now, use the Efron jackknife variant
robust<-diag(sqrt(hccm(model, type="hc3")))
robust

# t-test
model$coefficients/robust








# do it with no correlation between the error and X
# Create a fake data set
X<-runif(200, min=-10, max=10)
u<-rnorm(200, mean=0, sd=runif(200, min=0.5, max=2))
y<-3+2*X+u


# run a standard model
model<-lm(y~X)
summary(model)
plot(model$residuals~X)

# now, use White's heteroskedasticity-robust standard errors
robust<-diag(sqrt(hccm(model, type="hc0")))
robust

# t-test
model$coefficients/robust

# now, use the Efron jackknife variant
robust<-diag(sqrt(hccm(model, type="hc3")))
robust

# t-test
model$coefficients/robust
```
